```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
from sklearn.model_selection import train_test_split
from sklearn import metrics
import xgboost as xgb
```

```python
data1=pd.read_csv("train.csv")
data1 = data1[data1['cancel'] >= 0]
data1
```

```python
#NAclean 
y=data1["cancel"]
X=data1.drop(["cancel","dwelling.type","year"],axis=1)
dummyx=pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(dummyx, y, test_size=0.2, random_state=1)
```

```python
# 
data=pd.read_csv("cleaned.csv")
y=data["cancel"]
X=data.drop(["cancel","zip.code"],axis=1)
dummyx=pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(dummyx, y, test_size=0.2, random_state=1)
```

```python
#
data_train = xgb.DMatrix(X_train, y_train) 
#data_balance_train = xgb.DMatrix(X_balance_train, y_balance_train)
# XGBoost
data_test = xgb.DMatrix(X_test, y_test)
```

```python
#booster_04_raw 01010 'max_depth': 4, 'eta': 0.01
param = {'max_depth': 4, 'eta': 0.01, 'objective': 'binary:logistic'}#,'scale_pos_weight':1/scale}
watchlist = [(data_test, 'test'), (data_train, 'train')]
n_round = 1010
booster_04_raw= xgb.train(param, data_train, num_boost_round=n_round, evals=watchlist)
#imbalance
```

```python

pred = booster_04_raw.predict(data_test)
y = data_test.get_label()
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y, pred, pos_label=1)
metrics.auc(fpr, tpr)
```
